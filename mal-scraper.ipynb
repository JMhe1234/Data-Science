```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://myanimelist.net/topanime.php"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    
    table = soup.find("table", {"class": "top-ranking-table"})
    
    ranks = []
    titles = []
    scores = []
    
    for row in table.find_all("tr")[1:]:
        columns = row.find_all("td")
        if len(columns) == 5:
            rank = columns[0].get_text(strip=True)
            title = columns[1].get_text(strip=True)
            score = columns[2].get_text(strip=True)
            ranks.append(rank)
            titles.append(title)
            scores.append(score)
            
    anime = {"Rank": ranks, "Title": titles, "Score": scores}
    top_anime = pd.DataFrame(anime)
    
    top_anime.to_csv("top_animes_v1.csv", index=False)
    
    print("Data has been scraped and saved to 'top_animes.csv'.")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```
